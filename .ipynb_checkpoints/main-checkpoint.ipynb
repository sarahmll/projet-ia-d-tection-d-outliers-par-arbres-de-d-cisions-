```python
import pandas as pd
import seaborn as sns
import sklearn.cluster as cluster
import numpy as np

root = "./"
outlier_dataframe = pd.read_csv(root+'data.csv')
print (outlier_dataframe.head())
print (outlier_dataframe.info())
sns_plot=sns.pairplot(outlier_dataframe,hue ='Class')
sns_plot.savefig("output.png")

# Nombre d’instances dans le dataframe :
nb_lignes=outlier_dataframe.shape[0]
print(nb_lignes)
# Decompte du nombre d’instance de chaque classe dans le dataframe
series=outlier_dataframe['Class'].value_counts()
print(series)
outlier_dataframe=outlier_dataframe.rename(columns={'Item1': 1,'Item2':2})
```

```python
#FONCTION qui prend en parametres un datframe et et une liste d'attribut qu'on veut traiter
#et qui renvoie  le meilleur attribut et les 2 seuils a et b 
def getSplitParameters(donne,tabatt):
    tabatt=list(tabatt)
    ec1=np.std(donne[tabatt[0]])  
    a=0
    for i in range(1,len(tabatt)):
        ec2=np.std(donne[tabatt[i]])
        if(ec2>ec1):
            ec1=ec2
            a=i
    #calculs des seuils 
    dt=donne[tabatt[a]]
    mn=min(dt)
    mx=max(dt)
    ini=np.array([[mn],[mx]],np.float64)
    newdt = []
    for value in dt:
        newdt.append([value])
    kmeans=cluster.KMeans(n_clusters=2,init=ini,n_init=1)
    kmeans=kmeans.fit(newdt)
    s1,s2=kmeans.cluster_centers_
    return tabatt[a],s1[0],s2[0]
```

```python
def exactitudepondere(TN,FP,FN,TP):
    num1=TN/(TN+FP)
    num2=TP/(FN+TP)
    return (num1+num2)/2
def precision(TN,FP,FN,TP):
        return TP/(TP+FP)
def rappel(TN,FP,FN,TP):
    return TP/(FN+TP)
def exactitude(TN,FP,FN,TP):
    return (TN+TP)/(TN+FP+FN+TP)
```

```python
# une classe directdecision qui initialise un bolean (outlier) passé en parametre et un dataframe (dt) initialiser en NULL   
class directdecision:
    def __init__(self,outlier):
        if(outlier):
            self.outlier=0
        else:
            self.outlier=1
        self.dt=[]
```

```python
#   cette fonction permet d'evaluer un modele elle prend en parametres un dataframe et et le nom de l'attribut qui determine si une donnée est un outlier ou inlier   
# 
def evaluation(modele,indice_res):
    if(modele is None):
        return
    TN=FP=FN=TP=0
    if(type(modele)==directdecision):
        for index,row in modele.dt.iterrows() :
            if (modele.outlier==0):
                if(row[indice_res]==1):
                   
                    TP+=1
                else:
                    FP+=1
            else:
                if(row[indice_res]==1):
                    FN+=1
                else:
                    TN+=1
    else:
        TN,FP,FN,TP=evaluation(modele.L,indice_res)
        TNprim,FPprim,FNprim,TPprim=evaluation(modele.M,indice_res)
        TN+=TNprim
        FP+=FPprim
        FN+=FNprim
        TP+=TPprim
        TNprim,FPprim,FNprim,TPprim=evaluation(modele.R,indice_res)
        TN+=TNprim
        FP+=FPprim
        FN+=FNprim
        TP+=TPprim
    return TN,FP,FN,TP
```

```python
#decisionleaf est la classe qui represente notre feuille de decision on initialise un dataframe et un tableau passés en parametres 
# et initialise 3 directdecision L , M,Rqui réprensentent respetivement les données qui sont plus petites que le  seuil a donc des outliers ,
# ,les données qui sont les données plus grand que le seuil a et plus petit que le seuil b donc des inliers et les  données  
class decisionleaf:
    def __init__(self,donne,tabatt):
        self.donne=donne
        self.tabatt=tabatt
        self.att=0
        self.L=directdecision(True)
        self.M=directdecision(False)
        self.R=directdecision(True)
        self.curentatt ,self.a , self.b=getSplitParameters(donne,tabatt)
        Dl=donne[donne[self.curentatt]<self.a]
        Dm=donne[donne[self.curentatt]>=self.a]
        Dm=Dm[Dm[self.curentatt]<=self.b]
        Dr=donne[donne[self.curentatt]>self.b]
        self.L.dt=Dl
        self.M.dt=Dm
        self.R.dt=Dr
    def get_donne(self):
        return self.donne
    def get_tabatt(self):
        return self.tabatt
    
```

```python
# evaluation du premiere algo 4.2 avec une feuille
dl=decisionleaf(outlier_dataframe,{1, 2})
e=evaluation(dl,'Class')
print("la matrice de confusion du premiere algorithme est:")
print(str(e[0])+"  " +str(e[1]))
print(" "+str(e[2])+"  " +str(e[3]))
expond=exactitudepondere(e[0],e[1],e[2],e[3])
ex=exactitude(e[0],e[1],e[2],e[3])
pr=precision(e[0],e[1],e[2],e[3])
rap=rappel(e[0],e[1],e[2],e[3])
print("l'exactitude pondéré est de :"+ str(expond*100)+"%")
print("l'exactitude  est de :"+ str(ex*100)+"%")
print("la précision est de   :"+ str(pr*100)+"%")
print("le rappel est de :"+ str(rap*100)+"%")
```

```python
#cette classe permet de creer le noeud de notre arbre
class Node:
    def __init__(self , currentAttrib,a,b,L,M,R):
        self.currentAttrib=currentAttrib
        self.a=a
        self.b=b
        self.L=L
        self.M=M
        self.R=R

    
```

```python
# cette fonction  permet de construire l'arbre à partir d'un modéle donné 
def buildDecisionTree(D,central,attribIdx):
    cardD=D.shape[0]
    
    
    if (cardD>=4):
        cardattribidx=len(attribIdx)
        if(cardattribidx>=2):
            curentatt,a,b=getSplitParameters(D,attribIdx)
            Dl=D[D[curentatt]<a]
            Dm=D[D[curentatt]>=a]
            Dm=Dm[Dm[curentatt]<=b]
            Dr=D[D[curentatt]>b]
            attribIdx.remove(curentatt)
            L=buildDecisionTree(Dl,False,attribIdx)
            M=buildDecisionTree(Dm,True,attribIdx)
            R=buildDecisionTree(Dr,False,attribIdx)
            return Node(curentatt,a,b,L,M,R)
        else:
            return decisionleaf(D,attribIdx)
    else:
        if(central==True):
            d=directdecision(False)
            d.dt=D
            
        else:
            d= directdecision(True)
            d.dt=D
        return d
```

```python
# evaluation du deuxieme algo
tree=buildDecisionTree(outlier_dataframe, True, {1, 2})
e=evaluation(tree,'Class')
print("la matrice de confusion du deuxieme algorithme est:")
print(str(e[0])+"  " +str(e[1]))
print(" "+str(e[2])+"  " +str(e[3]))
print()
expond=exactitudepondere(e[0],e[1],e[2],e[3])
ex=exactitude(e[0],e[1],e[2],e[3])
pr=precision(e[0],e[1],e[2],e[3])
rap=rappel(e[0],e[1],e[2],e[3])
print("l'exactitude  est de :"+ str(ex*100)+"%")
print("l'exactitude pondéré est de :"+ str(expond*100)+"%")
print("la précision est de   :"+ str(pr*100)+"%")
print("le rappel est de :"+ str(rap*100)+"%")
```

```python
def buildFinalDecisionTree(D,central,attribIdx,hauteur,hauteurmax):
    cardD=D.shape[0]
    if (cardD>=4):
        if(hauteur<hauteurmax):
            curentatt,a,b=getSplitParameters(D,attribIdx)
            Dl=D[D[curentatt]<a]
            Dm=D[D[curentatt]>=a]
            Dm=Dm[Dm[curentatt]<=b]
            Dr=D[D[curentatt]>b]
            
            
            L=buildFinalDecisionTree(Dl,False,attribIdx,hauteur+1,hauteurmax)
            M=buildFinalDecisionTree(Dm,True,attribIdx,hauteur+1,hauteurmax)
            R=buildFinalDecisionTree(Dr,False,attribIdx,hauteur+1,hauteurmax)
            return Node(curentatt,a,b,L,M,R)
        else:
            return decisionleaf(D,attribIdx)
    else:
        if(central==False):
            d=directdecision(False)
            d.dt=D
            
        else:
            d= directdecision(True)
            d.dt=D
        return d
```

```python
# evaluation du troisieme algo
tree1=buildFinalDecisionTree(outlier_dataframe, True, {1, 2},1,1)
tree2=buildFinalDecisionTree(outlier_dataframe, True, {1, 2},1,2)
tree3=buildFinalDecisionTree(outlier_dataframe, True, {1, 2},1,3)
tree4=buildFinalDecisionTree(outlier_dataframe, True, {1, 2},1,4)
```

```python
# tableau contenant  l’ensemble des métriques
evtree1=evaluation(tree1,'Class')
evtree2=evaluation(tree2,'Class')
evtree3=evaluation(tree3,'Class')
evtree4=evaluation(tree4,'Class')
exact1=exactitude(evtree1[0],evtree1[1],evtree1[2],evtree1[3])
exact2=exactitude(evtree2[0],evtree2[1],evtree2[2],evtree2[3])
exact3=exactitude(evtree3[0],evtree3[1],evtree3[2],evtree3[3])
exact4=exactitude(evtree4[0],evtree4[1],evtree4[2],evtree4[3])
exactp1=exactitudepondere(evtree1[0],evtree1[1],evtree1[2],evtree1[3])
exactp2=exactitudepondere(evtree2[0],evtree2[1],evtree2[2],evtree2[3])
exactp3=exactitudepondere(evtree3[0],evtree3[1],evtree3[2],evtree3[3])
exactp4=exactitudepondere(evtree4[0],evtree4[1],evtree4[2],evtree4[3])
rap1=rappel(evtree1[0],evtree1[1],evtree1[2],evtree1[3])
rap2=rappel(evtree2[0],evtree2[1],evtree2[2],evtree2[3])
rap3=rappel(evtree3[0],evtree3[1],evtree3[2],evtree3[3])
rap4=rappel(evtree4[0],evtree4[1],evtree4[2],evtree4[3])
pr1=precision(evtree1[0],evtree1[1],evtree1[2],evtree1[3])
pr2=precision(evtree2[0],evtree2[1],evtree2[2],evtree2[3])
pr3=precision(evtree3[0],evtree3[1],evtree3[2],evtree3[3])
pr4=precision(evtree4[0],evtree4[1],evtree4[2],evtree4[3])

ar = np.array([[exact1,exactp1,rap1,pr1],[exact2,exactp2,rap2,pr2],[exact3,exactp3,rap3,pr3],[exact4,exactp4,rap4,pr4]])
df = pd.DataFrame(ar, index = ['hauteur max=1', 'hauteur max=2', 'hauteur max=3','hauteur max=4'], columns = ['exactitude', 'exactitudepondere', 'rappel', 'precision'])
print(df)
```
